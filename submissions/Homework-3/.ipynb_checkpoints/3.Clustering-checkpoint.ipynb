## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
loc = boston[['LONGITUDE','LATITUDE']]
loc = StandardScaler().fit_transform(loc)
plt.scatter(loc[:,0],loc[:,1], s = 1)
```

```python
#Code based/taken directly from Professor Terzi's ipython notebook on Clustering
kmeans = KMeans(init='k-means++', n_clusters=9, n_init=100)
kmeans.fit_predict(loc)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

print "The total error of the clustering is: ", error

colors = np.array(['#8e7618','#40e0d0','#009bff','#a200ff','#f47835','#99ccff','#7bf6b6','#ffaadd','#fb9800','#fcee21','#0a6308',
                   '#65457e','#f5be91','#b2b2b2','r'])
colors = np.hstack([colors] * 20)
plt.scatter(loc[:, 0], loc[:, 1], color=colors[labels].tolist(), s=1, alpha=0.8)
```

```python
error = np.zeros(20)
error[0] = 0;
for k in range(1,20):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=20)
    kmeans.fit_predict(loc)
    error[k] = kmeans.inertia_

plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
plt.ylabel('Error')
```

I chose k to be 9 in my clustering.  9 is the best choice for two reasons: the first is that after 9 the error rate decreases at a slightly slower rate (as seen in the graph above), and secondly it better represents the neighborhoods of Boston.  At k = 9 and k = 10 there is little difference above x = 0, it is below that makes a difference.  No matter how I clustered, Roxbury was impossible to outline because it is in the middle of other populous neighborhoods, so I attempted to focus on the other neighborhoods accepting the fact that Roxbury was swallowed up.  When there are 9 clusters there is a more accurate representation of the neighborhood boundaries, even if some of them are combined.

## Step 2. Using multiple features

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!) Filter the dataset to get more features, transform them as required and compute another clustering. You should argue about the quality of your clustering and use domain knowledge (i.e. the fact that you are locals) to explore the clusters you produced and present them to the reader. **(pts 30)**

```python
#Create an n by d matrix A where n is the date and d is a list of neighborhoods such that Ai,j represents the frequency a call
#request was on time on the ith day in the jth neighborhood
l = []
labels = []
oneday = pd.TimeGrouper(freq = "1D")
boston.index = boston.OPEN_DT
neighborhoods = boston.neighborhood.dropna().unique()
neighborhoods = np.delete(neighborhoods, [14,19,20,21]) #remove unused neighborhoods
ontime = boston[['OPEN_DT','OnTime_Status']]
for n in neighborhoods [0:18]:
    byday = boston[(boston.neighborhood == n) & (boston.OnTime_Status == 'ONTIME')].groupby(oneday).count()
    byday.rename(columns={'OnTime_Status':'OnTime_Status ' + n}, inplace=True)
    labels.append('OnTime_Status ' + n)
    byday_select = byday['OnTime_Status ' + n]
    l.append(byday_select)
ontime_n = pd.concat(l, axis = 1).fillna(value=0)
```

```python
ontime = StandardScaler().fit_transform(ontime_n)
U, s, V = np.linalg.svd(ontime)
```

```python
#taken from class notes
def svd_projection(data,k):
    U, s, V = np.linalg.svd(data)
    Uk = U[:,0:k]
    Vk = V[0:k,:]
    Sk = np.diag(s[0:k])
    datak = np.dot(Uk,Sk)
    return datak

svdX = svd_projection(ontime,5)
svdX_df = pd.DataFrame(svdX)
g = sns.PairGrid(svdX_df)
g.map(plt.scatter)
```

Visually, the groupings in this data all appear to be a clustering of 1.  This means that the service time in all areas of Boston is roughly the same.  There are outliers in the data, but the data is not filtered based off of service types.  For instance, during one of the snow storms one would expect to see a single small clustering with more outliers considering there would have been lots of requests for snow clearing and, inevitbly, some requests would have gotten delayed longer.  But this delay does not appear according to the data, to center around any one location.  It is not important that there is a single correlation among the timeliness of call response in any one location but the fact that there seems to be a similar correlation amongst all pair-wise categories.

```python
error = np.zeros(20)
error[0] = 0;
for k in range(1,20):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=20)
    kmeans.fit_predict(ontime)
    error[k] = kmeans.inertia_

plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
plt.ylabel('Error')
```

```python
k=3
kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=100, n_init=10, random_state=0)
kmeans.fit_predict(ontime)
centroidsk = kmeans.cluster_centers_
labelsk = kmeans.labels_
errork = kmeans.inertia_

print "Centroids: "
print centroidsk
print "Error : " + str(errork)
print "Data Point Labels: "
print labelsk
```

When we look into the actual large dataset we see that few clusterings are needed, as shown in the 2-d representation.  Three clusterings seems optimal because at four, you begin to have single nodes that form clusters and at two you have an error increase of around 100.  Although we see that most of the points are in the first cluster, which mirrors much of the behavior we saw when taking the data projections.  With few clusterings and this particular data set, we understand that the variation in service time per neighborhood is small.  The clusters that we are seeing now most likely refer to the excessive snow we've had.  When there were many storms together, the service time must have suffered and then in coming weeks when the city still needed to clean and repair, there must have been another clustering of different service times.  I believe the analysis of this data is positive and also expected, that no particular area of boston suffers particular delays but that snow changes service time for everyone.

## Step 3. Detecting outliers

Now that we have a clustering of the records in the dataset, we can detect the outlying ones. These are the ones who are the farthest from the centroids of their clusters. Track them down and describe any interesting observations that you can make. **(pts 20)** 

```python
#prints furthest points from the centroid for each cluster, a single outlier
outliers = []
for i in range(3):
    furthest = 0
    for l in range(len(labelsk)):
        #print "i : " + str(i) + " l : " + str(l)
        li = labelsk[l]
        if li == i:
            #print euclidean(centroidsk[i], ontime[l,:])
            #print euclidean(centroidsk[i], ontime[furthest,:])
            if euclidean(centroidsk[i], ontime[l,:]) > euclidean(centroidsk[i], ontime[furthest,:]):
                #print "greater than"
                furthest = l
    #print "furthest" + str(furthest)
    outliers.append(furthest)
    
for f in outliers:
    print ontime_n.iloc[[f]]
```

In the first cluster the furthest from the centroid is a day when there was a high volume of ontime requests, due to more requests which were due to snow.  The second cluster seems to have its furthest point be one where very few requests were ontime which was most likely due to a few number of requests. The third seems a mix of the two where some are much higher than others, although the average of these numbers tends to be lower.  This reenforces the idea that the first cluster is describing times when there wasn't much snow and less calls, the second cluster appears to be the time when there was the most snow and the third cluster is some period between the heavy influx of calls and the calmer days.  This proves that three clusters was the correct choice.  If we note the graph from the end of the last homework we see that there is not a lot of middling calls, there tend to be a lot or a little which would support a two-clustering paradigm but as in all data, there are no absolutes and the middle must be considered as it is in the third cluster.
